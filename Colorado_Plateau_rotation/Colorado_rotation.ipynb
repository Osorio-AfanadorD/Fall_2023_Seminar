## Geologic background of the Colorado Plateau and its Mesozoic (Cretaceous) to Cenozoic plate tectonics history

### Laramide orogeny
- 80-40 Ma (Dickinson et al., 1988)
- resulted in counter clockwise rotation of the Colorado Plateau
- caused by subduction of the Farallon plate under the North American plate
- hypothesized to be driven by basal traction during an interval of horizontal flat-slab subduction

#### Map of the North American Cordilleran system.
- The North American Cordillera orogenic system developed in response to subduction of oceanic (Farallon and Kula) plates and accretion of terranes along the North American plate margin during Jurassic to Paleogene time. 
- Relative convergence rates between the Farallon plate and the NA plate decreased to an estimated ~5 cm/yr from ~ 145–130 Ma, increased to 8–12 cm/yr from ~ 120–50 Ma, and then decreased after 50 Ma. 
- The direction of relative motion changed to SW–NE during the Late Cretaceous and Paleogene with dextral oblique convergence along much of the margin, which was greatest to the north where the Kula plate devel- oped at ~80 Ma. This dextral sense of shear is interpreted based on paleogeography reconstructions (magnetic anomaly?) and structural mapping (Tikoff and de Saint Blanquat, 1997) such as the Big Trail fault zone (Ver Ploeg and Greer, 1997). 
- Sevier fold and thrust belt; Laramide belt
- The Laramide orogeny is the Late Cretaceous to Paleocene (80 to 55 Ma) orogenic event that gave rise to the Laramide block uplifts in the United States, the Rocky Mountain fold-and-thrust belt in Canada and the United States, and the Sierra Madre Oriental fold-and-thrust belt in east-central Mexico. The Laramide orogeny is believed to post-date the Jurassic and late Early Cretaceous accretion of the terranes that make up much of the North American Cordillera, precluding a collisional origin for Laramide orogenesis. Instead, the deformation belt along much of its length likely developed 700–1500 km inboard of the nearest convergent margin.

### The approach of Bryan and Gordon

- The idea is based on that had the Colorado Plateau rotated as a rigid block, its apparent polar wander path would be systematically different from that of the cratonic North America, possibly through the Cretaceous. 

- A first-order approach is to compare poles of the same agee from on and off the plateau. Gordon 1984 noted systematic differences and hypothesized to be either small < 10° of rotation. 

- But poles of exactly the same age are hard to find, counting also that many are sedimentary poles. 

- One solution is to use many poles and simultaneously comparing all available paleoamgentic data from on and off the plateau

- using a maximum likelihood approach to estimate the amount of rotation using poles through a section of the apparent polar wander (on and off the plateau)

- The rotation of the Colorado Plateau that yields a smooth APW path with the smallest dispersionis the maximum likelihood estimate of the rotation. 

- confidence limits are determined using a F ratio test 

- **note that the compilation by Gordon 1984 uses a sigma values for the standard deviation of the poles instead of A95. This is because the analyses tries to propagate bedding tilt correction, inclination shallowing in sedimentary rocks, dual polarity magnetic overprint uncertainties. In the plots below, I simply devide their sigma values by 0.58 and derived a calculated A95 values associated with the poles. We could, and perhaps want to, recompile the site level data and do this with bootstrap or Monte Carlo approach.**

Gordon further used geodedic data to independently estimate the rotation of the Colorado Plateau.

#### make functions for small circle fits

```python
def decompose_XY(matrix):
    matrix = np.array(matrix)
    X = matrix[:, [0,1]]
    Y = matrix[:, 2]
    return X, Y

def get_ortho_fitting_plane_vector(coords):

    # Calculate the covariance matrix of the coordinates
    covariance_matrix = np.cov(coords, rowvar=False) # Variables = columns

    # Calculate the eigenvalues & eigenvectors of the covariance matrix
    e_val, e_vect = np.linalg.eig(covariance_matrix)

    # The normal vector to the plane is the eigenvector associated to the minimum eigenvalue
    min_eval = np.argmin(e_val)
    normal_vector = e_vect[:, min_eval]

    return normal_vector

def least_squares_by_solve(X, Y):
    return solve(X.T @ X, X.T @ Y)

def add_ones_column(X):
    return np.hstack([np.ones((X.shape[0],1)), X])

def model_force_origin(X, Y):
    # this function forces the result to be a plane passing through the origin 
    # the result circle will be a great circle
    
    theta_hat = least_sqaures_by_solve(X, Y)
    return X @ theta_hat

def model_append_ones(X, Y):
    # this function returns a plane parameters of best fit to the data
    # the return values match (D, A, B)
    X_append_ones = add_ones_column(X)
    theta_hat = least_squares_by_solve(X_append_ones, Y)
    return X_append_ones @ theta_hat

def get_rho(A, B, C, D):
    # plane equation in form of Ax+By+Cz = D
    return -D/np.sqrt(A**2+B**2+C**2)

def get_circle_center(A, B, C, D):
    # plane equation in form of Ax+By+Cz = D
    rho = get_rho(A, B, C, D)
    center = rho/np.sqrt(A**2+B**2+C**2)*np.array([A, B, C])
    return center

def make_small_circle(rho, k):
    lon = np.linspace(0, 2*np.pi, 100)
    lat = (np.arctan(k*np.sin(lon))-np.arcsin(rho*np.sqrt(1+k**2)/(1+(k*np.sin(lon))**2)))*180/np.pi
    lon = lon*180/np.pi
    
    lon_lat = np.array([lon, lat]).T
    lon_lat_cart = pmag.dir2cart(lon_lat)
    
    return lon_lat_cart
    
def rodrigues_rot(P, n0, n1):
    # rotate a series of points given a rotation defined by starting vector n0 and ending vector n1
    
    # If P is only 1d array (coords of single point), fix it to be matrix
    if P.ndim == 1:
        P = P[np.newaxis,:]
    
    # Get vector of rotation k and angle theta
    n0 = n0/np.linalg.norm(n0)
    n1 = n1/np.linalg.norm(n1)
    k = np.cross(n0,n1)
    k = k/np.linalg.norm(k)
    theta = np.arccos(np.dot(n0,n1))
    
    # Compute rotated points
    P_rot = np.zeros((len(P),3))
    for i in range(len(P)):
        P_rot[i] = P[i]*np.cos(theta) + np.cross(k,P[i])*np.sin(theta) + k*np.dot(k,P[i])*(1-np.cos(theta))

    return P_rot

def kappa_from_alpha95(alpha95):
    return 140. * 140. / alpha95 / alpha95

def fit_small_circle_by_ortho(points, return_dir = True):

    ABC = get_ortho_fitting_plane_vector(points)
    A, B, C = ABC[0], ABC[1], ABC[2]
    P_bar = np.mean(points, axis=0)
    D = np.dot(-P_bar, ABC)
    
    rho = get_rho(A, B, C, D)
    circle_center = get_circle_center(A, B, C, D)
    circle_center_lon_lat = pmag.cart2dir(circle_center)
    
    initial_circle_cart = make_small_circle(rho, 0)
    rotated_circle_cart = rodrigues_rot(initial_circle_cart, np.array([0,0,1]), circle_center)

    rotated_circle_dir = pmag.cart2dir(rotated_circle_cart)

    rotated_circle_lon, rotated_circle_lat = rotated_circle_dir.T[0], rotated_circle_dir.T[1]
    
    if return_dir == True:
        return [rotated_circle_lon, rotated_circle_lat], circle_center_lon_lat
    else:
        return rotated_circle_cart
```

```python
# make cartesian coordinates of all poles
pole_carts = pmag.dir2cart(np.array([Gordon1984_table_1['pole longitude'], Gordon1984_table_1['pole latitude']]).T)

fit_small_circle, euler_pole = fit_small_circle_by_ortho(pole_carts)
```
