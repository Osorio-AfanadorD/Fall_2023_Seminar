```python
import pmagpy.ipmag as ipmag
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
Kulakov_data = pd.read_csv('Kulakov2021_sites.csv', sep=',', header=0)
```

```python
Kulakov_data
```

```python
ipmag.plot_net(1)
ipmag.plot_di(Kulakov_data['Dg'], Kulakov_data['Ig'])
plt.title('geographic data from Kulakov et al. (2021)')
plt.show()

ipmag.plot_net(1)
ipmag.plot_di(Kulakov_data['Ds'], Kulakov_data['Is'])
plt.title('tilt corrected data from Kulakov et al. (2021)')
plt.show()
```

```python
Kulakov_mean_pole = ipmag.fisher_mean(Kulakov_data['Plong'],Kulakov_data['Plat'])
ipmag.print_pole_mean(Kulakov_mean_pole)
```

```python
pole_map = ipmag.make_orthographic_map(central_latitude=90)
ipmag.plot_vgp(pole_map,Kulakov_data['Plong'],Kulakov_data['Plat'],
               color='pink',markersize=30)
ipmag.plot_pole(pole_map,Kulakov_mean_pole['dec'],Kulakov_mean_pole['inc'],Kulakov_mean_pole['alpha95'],
                marker='s',color='red',markersize=50)
```
